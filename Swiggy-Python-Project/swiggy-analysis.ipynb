```python
import pandas as pd
```

```python
df = pd.read_excel("../data/Swiggy Raw Data analysis.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Restaurant Name'].value_counts().head(10)
```

```python
df['City'].value_counts()
```

```python
df['Price (INR)'].mean()
```

```python
df.sort_values(by='Rating', ascending=False)[['Restaurant Name','Rating']].head(10)
```

```python
import matplotlib.pyplot as plt

top_cities = df['City'].value_counts().head(5)

top_cities.plot(kind='bar', figsize=(8,5))

plt.title("Top 5 Cities by Orders")
plt.xlabel("City")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

plt.show()
```

```python
print("Insight:")
print("Bengaluru has the highest number of orders, almost double compared to other cities.")
print("Other cities like Mumbai, Hyderabad, Jaipur, and Lucknow have similar order volumes.")
```

```python
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['Price (INR)'])

plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Price (INR)'], y=df['Rating'])

plt.title("Price vs Rating")
plt.xlabel("Price")
plt.ylabel("Rating")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['Price (INR)'], bins=50)

plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Price (INR)'], bins=50)

plt.xlim(0, 1000)  # focus on main data
plt.title("Price Distribution (0–1000 Range)")
plt.xlabel("Price")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x=df['Price (INR)'], 
    y=df['Rating'],
    alpha=0.3   # transparency
)

plt.title("Price vs Rating")
plt.xlabel("Price")
plt.ylabel("Rating")

plt.show()
```

```python
print("Insights:")
print("1. Most food items are priced below ₹500.")
print("2. There are few high-price outliers up to ₹8000.")
print("3. Ratings are mostly between 3.5 to 5.")
print("4. No strong correlation between price and rating.")
```
